```python
# しゃくとり法
# Subsequence

from bisect import bisect_left

n = int(input())
S = int(input())
a = list(map(int,input().split()))

sum = [0]*(n+1)

# O(n log n)
def solve1():
    for i in range(n):
        sum[i+1] = sum[i]+a[i]
    if sum[n] < S:
        print(0)
        return
    res = n
    s = 0
    while sum[s]+S <= sum[n]:
        t = bisect_left(sum,sum[s]+S)
        res = min(res,t-s)
        s += 1
    print(res)
    
# O(n)
def solve2():
    res = n+1
    s = 0
    t = 0
    tmp = 0
    for _ in range(n):
        while t < n and tmp < S:
            tmp += a[t]
            t += 1
        if tmp < S:
            break
        res = min(res,t-s)
        tmp -= a[s]
        s += 1
    if res > n:
        res = 0
    print(res)

"""
input1
10
15
5 1 3 5 10 7 4 9 2 8

output1
2

input2
5
11
1 2 3 4 5

output2
3
"""
solve1()
solve2()
```

```python
# Jessica's Reading Problem

P = int(input())
a = list(map(int,input().split()))
b = list(set(a))

def solve():
    n = len(b)
    s = 0
    t = 0
    num = 0
    res = P
    count = [0 for i in range(max(a)+2)]
    while True:
        while t < P and num < n:
            if count[a[t]] == 0:
                num += 1
            count[a[t]] += 1
            t += 1
        if num < n:
            break
        res = min(res,t-s)
        if count[a[s]]-1 == 0:
            num -= 1
        count[a[s]] -= 1
        s += 1
    print(res)
            
"""
input1
5
1 8 8 8 1

output1
2
"""

solve()
```

```python
# 反転
# Fence The Right Way

N = int(input())
d = list(input())
d = [1 if i == "B" else 0 for i in d]

def cal(n,d,k):
    f = [0 for i in range(N)]
    res = 0
    s = 0
    for i in range(n-k+1):
        if (d[i]+s)%2 != 0:
            res += 1
            f[i] = 1
        s += f[i]
        if i - k + 1 >= 0:
            s -= f[i-k+1]
    for i in range(n-k+1,n):
        if (d[i]+s)%2 != 0:
            return -1
        if i - k + 1 >= 0:
            s -= f[i-k+1]
    return res
def solve():
    kk = 1
    mm = N
    for k in range(1,N+1):
        m = cal(N,d,k)
        if m >= 0 and mm > m:
            kk = k
            mm = m
    print(kk,mm)

"""
input1
7
BBFBFBB

output1
3
3
"""

solve()
```

```python
# 弾性衝突
# Physics Experiment

from decimal import Decimal, ROUND_HALF_UP, ROUND_HALF_EVEN

N = int(input())
H = int(input())
R = int(input())
T = int(input())

g = 10.0

def cal(N,H,R,T,g):
    if T < 0:
        return H
    t = (2*H/g)**0.5
    k = T//t
    if k%2 == 0:
        d = T-k*t
        return H-g*d*d/2.0
    else:
        d = k*t+t-T
        return H-g*d*d/2.0

def solve():
    y = [-1 for i in range(N)]
    y = [cal(N,H,R,T-i,g) for i in range(N)]
    y.sort()
    for i in range(N):
        print(Decimal(str(y[i]+2*R*i/100.0)).quantize(Decimal('0.01'), rounding=ROUND_HALF_UP))

"""
input1
1
10
10
100

output1
4.95

input2
2
10
10
100

output2
4.95 10.20
"""

solve()
```
