探索のメモ化と動的計画法

```python
# 01ナップサック問題
# 全探索

n = int(input())
w = []
v = []
for _ in range(n):
    tmp1,tmp2 = map(int,input().split())
    w.append(tmp1)
    v.append(tmp2)
W = int(input())

def rec(i,j):
    res = 0
    if i == n:
        res = 0
    elif j < w[i]:
        res = rec(i+1,j)
    else:
        res = max(rec(i+1,j),rec(i+1,j-w[i])+v[i])
    return res
def solve():
    print(rec(0,W))

"""
input1
4
2 3
1 2
3 4
2 2
5

output1
7
"""

solve()
```

```python
# 01ナップサック問題
# 全探索メモ化

n = int(input())
w = []
v = []
for _ in range(n):
    tmp1,tmp2 = map(int,input().split())
    w.append(tmp1)
    v.append(tmp2)
W = int(input())

INF = -1
dp = [[INF]*(W+1) for _ in range(n+1)]

def rec(i,j):
    if dp[i][j] >= 0:
        return dp[i][j]
    res = 0
    if i == n:
        res = 0
    elif j < w[i]:
        res = rec(i+1,j)
    else:
        res = max(rec(i+1,j),rec(i+1,j-w[i])+v[i])
    dp[i][j] = res
    return res
def solve():
    print(rec(0,W))

"""
input1
4
2 3
1 2
3 4
2 2
5

output1
7
"""
    
solve()
```

```python
# 01ナップサック問題
# dp問題化(1)

n = int(input())
w = []
v = []
for _ in range(n):
    tmp1,tmp2 = map(int,input().split())
    w.append(tmp1)
    v.append(tmp2)
W = int(input())

INF = 0
dp = [[INF]*(W+1) for _ in range(n+1)]

def solve():
    for i in range(n-1,-1,-1):
        for j in range(W+1):
            if j < w[i]:
                dp[i][j] = dp[i+1][j]
            else:
                dp[i][j] = max(dp[i+1][j],dp[i+1][j-w[i]]+v[i])
    print(dp[0][W])

"""
input1
4
2 3
1 2
3 4
2 2
5

output1
7
"""
    
solve()
```

```python
# 最長共通部分問題

n = int(input())
m = int(input())
s = input()
t = input()

INF = 0
dp = [[INF]*(m+1) for _ in range(n+1)]

def solve():
    for i in range(n):
        for j in range(m):
            if s[i] == t[j]:
                dp[i+1][j+1] = dp[i][j] + 1
            else:
                dp[i+1][j+1] = max(dp[i][j+1],dp[i+1][j])
    print(dp[n][m])

"""
input1
4
4
abcd
becd

output1
3
"""
    
solve()
```

```python
# 個数制限なしナップサック問題
# 3重ループ

n = int(input())
w = []
v = []
for _ in range(n):
    tmp1,tmp2 = map(int,input().split())
    w.append(tmp1)
    v.append(tmp2)
W = int(input())

INF = 0
dp = [[INF]*(W+1) for _ in range(n+1)]

def solve():
    for i in range(n):
        for j in range(W+1):
            k = 0
            while k*w[i] <= j:
                dp[i+1][j] = max(dp[i+1][j],dp[i][j-k*w[i]]+k*v[i])
                k += 1
    print(dp[n][W])

"""
input1
3
3 4
4 5
2 3
7

output1
10
"""
    
solve()
```

```python
# 01ナップサック問題その2

n = int(input())
w = []
v = []
for _ in range(n):
    tmp1,tmp2 = map(int,input().split())
    w.append(tmp1)
    v.append(tmp2)
W = int(input())

INF = 10**8
dp = [[INF]*(n*max(v)+1) for _ in range(n+1)]
dp[0][0] = 0

def solve():
    for i in range(n):
        for j in range(n*max(v)+1):
            if j < v[i]:
                dp[i+1][j] = dp[i][j]
            else:
                dp[i+1][j] = min(dp[i][j],dp[i][j-v[i]]+w[i])
    res = 0
    for i in range(n*max(v)+1):
        if dp[n][i] <= W:
            res = i
    print(res)

"""
input1
4
2 3
1 2
3 4
2 2
5

output1
7
"""
    
solve()
```

```python
# 個数制限付き部分和問題1

n = int(input())
a = list(map(int,input().split()))
m = list(map(int,input().split()))
K = int(input())

INF = 0
dp = [[INF]*(K+1) for _ in range(n+1)]

def solve():
    dp[0][0] = True
    for i in range(n):
        for j in range(K+1):
            k = 0
            while k <= m[i] and k * a[i] <= j:
                dp[i+1][j] |= dp[i][j-k*a[i]]
                k += 1
    if dp[n][K]:
        print("Yes")
    else:
        print("No")

"""
input1
3
3 5 8
3 2 2
17

output1
Yes
"""
        
solve()
```

```python
# 個数制限付き部分和問題2

n = int(input())
a = list(map(int,input().split()))
m = list(map(int,input().split()))
K = int(input())

INF = -1
dp = [INF]*(K+1)

def solve():
    dp[0] = 0
    for i in range(n):
        for j in range(K+1):
            if dp[j] >= 0:
                dp[j] = m[i]
            elif j < a[i] or dp[j-a[i]] <= 0:
                dp[j] = -1
            else:
                dp[j] = dp[j-a[i]]-1
    if dp[K] >= 0:
        print("Yes")
    else:
        print("No")

"""
input1
3
3 5 8
3 2 2
17

output1
Yes
"""

solve()
```

```python
# 最長増加部分列問題1

n = int(input())
a = list(map(int,input().split()))

INF = -1
dp = [INF]*n

def solve():
    res = 0
    for i in range(n):
        dp[i] = 1
        for j in range(i):
            if a[j] < a[i]:
                dp[i] = max(dp[i],dp[j]+1)
        res = max(res,dp[i])
    print(res)
    
"""
input1
5
4 2 3 1 5

output1
3
"""
    
solve()
```

```python
# 最長増加部分列問題2(解決したよ〜)
# https://qiita.com/ta7uw/items/d6d8f0ddb215c3677cd3

import bisect
import numpy as np

n = int(input())
a = list(map(int,input().split()))

INF = 0
dp = np.array([INF]*(n+1))

def solve():
    dp[0] = a[0]
    length = 1
    for i in range(1,n):
        if dp[length-1] < a[i]:
            dp[length] = a[i]
            length += 1
        else:
            index = bisect.bisect_left(dp,a[i],0,length)
            dp[index] = a[i]
    return print(length)

"""
input1
5
4 2 3 1 5

output1
3
"""

solve()
```

```python
# 分割数

n = int(input())
m = int(input())
M = int(input())

INF = 10**8
dp = [[INF]*(n+1) for _ in range(m+1)]

def solve():
    dp[0][0] = 1
    for i in range(m+1):
        for j in range(n+1):
            if j - i >= 0:
                dp[i][j] = (dp[i-1][j]+dp[i][j-i])%M
            else:
                dp[i][j] = dp[i-1][j]
    print(dp[m][n])

"""
input1
4
3
10000

output1
4
"""
    
solve()
```

```python
# 重複組み合わせ

n = int(input())
m = int(input())
a = list(map(int,input().split()))
M = int(input())

INF = 0
dp = [[INF]*(M+1) for _ in range(n+1)]

def solve():
    for i in range(n+1):
        dp[i][0] = 1
    for i in range(n):
        for j in range(m+1):
            if  j-1-a[i] >= 0:
                dp[i+1][j] = (dp[i+1][j-1] + dp[i][j] - dp[i][j-1-a[i]] + M)%M
            else:
                dp[i+1][j] = (dp[i+1][j-1] + dp[i][j])%M
    print(dp[n][m])

"""
input1
3
3
1 2 3
10000

output1
6
"""
    
solve()
```
